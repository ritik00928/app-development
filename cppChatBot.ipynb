```python
#data loading 
from langchain_community.document_loaders import TextLoader

loader = TextLoader("C:/Users/asus/OneDrive/Desktop/gen AI/cppTextData.txt", encoding="utf-8")
txtdata = loader.load()
print(txtdata)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
final_documents = text_splitter.split_documents(txtdata)
final_documents
```

```python
# data embedding means converting txt data into vectors
from langchain_huggingface import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
embedding
```

```python
# Vector Store
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(final_documents, embedding)
db
```

```python
# Retrieval similarity search
query = input("Enter your question for C++ (I am your assistant): ")

docs = db.similarity_search(query)

for i in range(len(docs)):
    print(docs[i].page_content)
```
